Create a predictor matrix with variables of your choice. State your reason.

```python
# ANSWER
#X = original_data.drop(['alcohol', 'residual sugar', 'fixed acidity', 'volatile acidity', 'citric acid'],axis=1).values
X = original_data.drop(['chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'quality', 'red_wine', 'fixed acidity'],axis=1)
X
```

### 4. Using Linear Regression Create a Model and Test Score

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Create a model for Linear Regression
reg = LinearRegression()

# Fit the model with the Training data
reg.fit(X_train, y_train)

# Make predictions using the testing set
y_pred_test = reg.predict(X_test)
y_pred_train = reg.predict(X_train)

# Calculate the score (R^2 for Regression) for Training Data
print('Coefficient of determination: %f \n'
      % r2_score(y_train, y_pred_train))

# Calculate the score (R^2 for Regression) for Testing Data
print('Coefficient of determination: %f \n'
      % r2_score(y_test, y_pred_test))

# The coefficients
print('Coefficients: \n', reg.coef_)
# The mean squared error
print('\nMean squared error: %f'
      % mean_squared_error(y_test, y_pred_test))

print(y_pred_test)
print(y_pred_train)
```

### BONUS: Cross validation

```python
# Cross validation 
from sklearn.model_selection import KFold 
from sklearn.metrics import mean_squared_error
```

```python
# Set up 5-fold cross validation  
k_fold = KFold(5, shuffle=True)
train_scores = []
train_rmse = []
test_scores = []
test_rmse = []

for k, (train, test) in enumerate(k_fold.split(X)):
    
    # Get training and test sets for X and y
    X_train = X.iloc[train, ]
    y_train = y.iloc[train, ]
    X_test = X.iloc[test, ]
    y_test = y.iloc[test, ]
    
    # Fit model with training set
    reg.fit(X_train, y_train)
    
    # Make predictions with training and test set
    y_pred_test = reg.predict(X_test)
    y_pred_train = reg.predict(X_train)
    
    # Score R2 and RMSE on training and test sets and store in list
    
    # Calculate the score (R^2 for Regression) for Training Data
    train_scores.append(r2_score(y_train, y_pred_train))

    # Calculate the score (R^2 for Regression) for Testing Data
    test_scores.append(r2_score(y_test, y_pred_test))
    
    # The mean squared error - test
    test_rmse.append(mean_squared_error(y_test, y_pred_test))
    
    # The mean squared error - train
    train_rmse.append(mean_squared_error(y_train, y_pred_train))
    
    
# Create a metrics_df dataframe to display r2 and rmse scores
metrics_df = pd.DataFrame(list(zip(train_scores, test_scores, test_rmse, train_rmse)),columns =['train_scores', 'test_scores','test_rmse','train_rmse'])
metrics_df
```
