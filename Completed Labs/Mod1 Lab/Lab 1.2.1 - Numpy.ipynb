# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

# get matplot to show charts inline
%matplotlib inline
sns.set()
````

````python
np.random.seed(0)
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

**1. Find Dimension of Array**

Find dimension of each array: height, weight, population
```
# Expected Ouput
Dimension(height): 1
Dimension(weight): 1
Dimension(population): 2
```

````python
print('Dimension(height): ', np.ndim(height))
print('Dimension(weight): ', np.ndim(weight))
print('Dimension(population): ', np.ndim(population))
````

**2. Find Shape of Array**

Find shape of each array: height, weight, population
```
# Expected Ouput
Shape(height) : (7000,)
Shape(weight) : (7000,)
Shape(population) : (7000, 2)
```



````python
print('Shape(height): ', np.shape(height))
print('Shape(weight): ', np.shape(weight))
print('Shape(population): ', np.shape(population))
````

**3. Find Size of Array**

Find shape of each array: height, weight, population
```
# Expected Ouput:

Size(weight) : 7000

```



````python
print('Size(height): ', height.size)
print('Size(weight): ', weight.size)
print('Size(population): ', population.size)
````

**8. Create a histogram of height and weight**

````python
#help(plt.hist)
plt.figure(figsize=(20,10))
plt.hist(height, color="green", bins=40, alpha = 0.3)
plt.xlabel("Height", fontsize=25)
plt.show()
````

````python
plt.figure(figsize=(20,10))
plt.hist(weight, color="red", bins=40, alpha = 0.3)
plt.xlabel("Weight", fontsize=25)
plt.show()
````

**9. Create a boxplot of height and weight**

````python
#help(plt.boxplot)
plt.figure(figsize=(20,10))
plt.boxplot(height)
plt.show()
````

````python
plt.figure(figsize=(20,10))
plt.boxplot(weight)
plt.show()
````

**10. Create scatterplot of height and weight**

````python
#help(plt.scatter)
plt.figure(figsize=(20,10))
plt.scatter(height, weight, marker='.', c='green', alpha = 0.2)
plt.xlabel("height (in Metre)", fontsize=25)
plt.ylabel("Weight (in kg)", fontsize=25)
plt.show()
````
